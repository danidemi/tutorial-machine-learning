```python


# Selection based on single predicate
firstclass = titanic[ titanic["pclass"]==1 ]
firstclass.shape[0]
```

```python
def fixAge(strAge):
    try:
        val = float(strAge)
    except:
        val = float("NaN")
    return val


# Selection based on multiple predicate
titanic["age_value"] = titanic["age"].apply( fixAge )
firstclass_and_young = titanic[ (titanic["pclass"]==1) & (titanic["age_value"] <= 6) ]
firstclass_and_young
```

```python
firstclass.age.unique()
```

```python
# let's select only the the first class with a known age
firstclass = firstclass[ firstclass["age"]!="?" ]
firstclass
```

```python
idx_1st_class = (titanic["pclass"]==1)
firstclass = titanic[ idx_1st_class ]
idx_with_age = firstclass["age"]!="?"
firstclass = firstclass[ idx_with_age ]
firstclass.age.unique()
```

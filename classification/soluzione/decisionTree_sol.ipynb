ESERCIZIO: utilizzando il dataset iris fai il training del modello utilizzando tutte le features a disposizione e disegna l'albero decisionale scegliendo come max_depth = 3 soluzione.

In base a come funziona il CART Algorithm, sai dire perchè l'albero ha esattamente quella struttura?

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()

iris.feature_names
```

```python
X = iris.data
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 3)
```

```python
tree_clf.fit(X,y)
```

```python
from sklearn.tree import export_graphviz 
import os
import graphviz

dot_data = export_graphviz(tree_clf, out_file=None, 
                      feature_names=iris.feature_names,  
                      class_names=iris.target_names,  
                      filled=True)  
graph = graphviz.Source(dot_data)  
graph 
```

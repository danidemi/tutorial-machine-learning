## ESERCIZIO: 
1. usando la funzione make_blobs(usa random_state=6) genera dei dati associati a 3 classi diverse
2. fai il training del modello SVC con kernel lineare;
3. genera il grafico dei punti e dei support_vectors;
4. fai predizioni sulla classe relativa al punto (-3,1) e controlla i valori della decision function;

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs


#Creazione di 100 punti divisi in tre classi
X, y = make_blobs(n_samples=100, centers=3, random_state=6)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
y
```

```python
# Facciamo il fit del modello: più C è alto più ci avviciniamo ad un hard margin classification
# decision_function_shape = 'ovr' -> avremo 10 algoritmi ognuno in grado di distinguere classi diverse 0 - non 0
# 1 - non 1 , ...
# y_train_5 = (y_train == 5)
# y_train_1 = (y_train == 1)
# y_train_0 = (y_train == 0)

clf = svm.SVC(C=1, decision_function_shape = 'ovr')
clf.fit(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=30)

plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

# Aggiungo nel grafico il punto di coordinate (-3,1)
plt.scatter([-3], [1], edgecolors='b', s = 100)
plt.show()
```

```python
clf.predict([[6,-10]])
```

```python
clf.decision_function([[6,-10]])
```
